What is CRUD?

In the context of data storage, CRUD stands for:
CREATE - Create new record.  INSERT in SQL, or POST in http.
READ   - Read or view a record.  SELECT in SQL, or GET in http.
UPDATE - Updates an existing record.  UPDATE in SQL, or PUT/PATCH in http.
DELETE - Deletes an existing record.  DELETE in SQL or http.
